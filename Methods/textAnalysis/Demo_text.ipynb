```python
import nltk
import urllib.request
import math
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer


#nltk.download()
```

```python
text = open('C:\\Users\\olliv\\Documents\\cogsys\\speeches\\2009-Obama.txt','r').read()
print(text)
```

```python

tokens = text.split()
tokens=[token.lower() for token in tokens if token.isalpha()]
print(tokens)
```

```python
freq = nltk.FreqDist(tokens) 
for key,val in freq.items(): 
    print ("%30s %5d" % (str(key), val))
```

```python
sw = stopwords.words('english')
print(sw)
```

```python
ctokens = tokens
for token in tokens:
    if token in sw:
        ctokens.remove(token)
print(ctokens)
```

```python
# Stem
#stemmer = PorterStemmer()
#for i in range(len(ctokens)):
#    ctokens[i] = stemmer.stem(ctokens[i])
    
# Lemmatize
lemmatizer = WordNetLemmatizer()
for i in range(len(ctokens)):
    ctokens[i] = lemmatizer.lemmatize(ctokens[i])

print(ctokens)
```

```python
freq = nltk.FreqDist(ctokens)
for key,val in freq.items(): 
    print ("%30s %5d" % (str(key), val))
```

```python
pos = open('C:\\Users\\olliv\\Documents\\cogsys\\positive-words.txt','r').read().split()
print(pos)
```

```python
neg = open('C:\\Users\\olliv\\Documents\\cogsys\\negative-words.txt','r').read().split()
print(neg)
```

```python
countpositive = countnegative = countneutral = counttotal = 0
for token in ctokens:
    counttotal = counttotal + 1
    cat = ""
    if token in pos:
        cat = cat + "POS"
        countpositive = countpositive + 1
    elif token in neg:
        cat = cat + "NEG"
        countnegative = countnegative + 1
    else:
        countneutral = countneutral + 1
    print("Word:", token, cat)
    
```

```python
print("Positive=%.3f Negative=%.3f Neutral=%.3f" % (countpositive/counttotal, countnegative/counttotal, countneutral/counttotal))

print("Positiveness=%.3f Affection=%.3f" % (math.log(countpositive/countnegative), 1-countneutral/counttotal))
```
