```python
import nltk
import urllib.request
import math
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
import numpy as np
```

```python
text = open('White_Fang.txt','r').read().lower()
text = text.replace('--',' ')
text = text.replace('.','')
text = text.replace(',','')
print(text)
```

```python
raws = text.split('\n')
print(raws[0:20])
```

```python
sw = stopwords.words('english')
pos = open('positive-words.txt','r').read().split()
neg = open('negative-words.txt','r').read().split()
```

```python
#remowing plank raws
for raw in raws :
    if raw == '':
        raws.remove(raw)

#The rest of text processing takes place inside the for loop
# x variable is only to limit ammount of printed raws
x=0
score = [None] * len(raws)
for raw in raws:
    #spliting each raw to separate words
    tokens = raw.split()
    
    #remowing stopwords
    atokens = tokens[:]
    btokens = tokens[:]
    for token in atokens:
        if token in sw:
            btokens.remove(token)
    
    # Lemmatize
    lemmatizer = WordNetLemmatizer()
    for i in range(len(btokens)):
        btokens[i] = lemmatizer.lemmatize(btokens[i])
    #sentiment score
    p=0
    n=0
    for i in range(len(btokens)):
        if btokens[i] in pos:
            p = p + 1
        elif btokens[i] in neg:
            n = n + 1
    score[x] = p-n
    
    if x < 10 :
        print("%80s %5d" % (raw, score[x]))
    x = x+1
```

we need [:] becouse otherwise program overwrite tokens. we need atokens and btokens becouse without any of theme the list from wchich we take the token will change while iterations. 
With out those things usualy the list was not clear from stopwords

```python
mean = [None]*len(score)
window = 100
for i in range(int(window/2),len(score)-int(window/2)):
    s=0
    for t in range(i-int(window/2),i+int(window/2)):
        s = s + score[t]
    mean[i] = s/window
```

Compering uotput of this cell and cell above we can see that such words as 'dog' and 'dogs' now are both 'dog'
