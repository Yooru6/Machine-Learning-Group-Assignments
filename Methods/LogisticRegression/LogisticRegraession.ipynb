```python
#importing necesery libraries
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.model_selection import cross_val_score
```

```python
# Defining names of columns of data set
names = ['pelvic incidence', 'pelvic tilt', 'lumbar lordosis angle', 
         'sacral slope', 'pelvic radius', 'grade of spondylolisthesis','label']

# read data
df = pd.read_csv('column_2C.dat', 
                 sep=' ',
                 names=names,
                     header=None, 
                     index_col = None, 
                     na_values = '?')
#There are no missing values
#Shuffling data ser
df = df.sample(frac = 1)

#printing shape and first 10 rows of data set
print(df.shape)
df.head(10)
```

```python
#Describing basic features of dataset
df.describe()
```

```python
#dividing into input and labels
X = df.iloc[:,:6]
Y = df.iloc[:,6]
```

```python
# build and fit model
reg = LogisticRegression(solver = 'lbfgs')
reg.fit(X,Y)
```

```python
# cross-validate
k = 10
scores = cross_val_score(estimator=reg,
                        X=X,
                        y=Y,
                        scoring="accuracy",
                        cv=k,
                        )
acc = scores.mean()
print("Accuracy calculated using %d-fold cross validation = %.3f" % (k, acc))
```

```python
print('The program will show you prediction if the patient has abnormalities in vertebral column or no.\n', 
       'The accuracy of prediction is %.3f ' % acc)
cont = 'yes'
while (cont == 'yes'):
    I = np.array([[0,0,0,0,0,0]])
    print('\nPlase enter the values of folowing parameters:\n')
    I[0,0] = float(input(names[0] + ' '))
    I[0,1] = float(input(names[1] + ' '))
    I[0,2] = float(input(names[2] + ' '))
    I[0,3] = float(input(names[3] + ' '))
    I[0,4] = float(input(names[4] + ' '))
    I[0,5] = float(input(names[5] + ' '))

    ypred = reg.predict(I)
    if ypred == 'AB':
        print(  '*********************************************',
              '\nPatient has abnormalities in vertebral column\n',
              '\n*********************************************')
    else:
        print('\n******************',
              '\nPatient is healthy\n',
              '\n******************')
    cont = input('Do you want to ask about next patient?[yes/no] ')
    
```
